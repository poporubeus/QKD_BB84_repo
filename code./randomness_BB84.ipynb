```python
def key_string(dimension):
  key = np.random.randint(2,size=100)
  return key

import matplotlib.pyplot as plt 
plt.style.use("seaborn-darkgrid")
import numpy as np

N=100
nbits = np.linspace(1,99,N)
key_bits = key_string(N)
fig=plt.figure(figsize=(20,3))
plt.plot(nbits,key_bits,marker='.',color='mediumvioletred',label='Alice mex gen')
plt.xlabel=('Number of bits')
plt.yaxis=('Value')
plt.title('Randomness')
plt.legend(loc=4)
#plt.savefig('Randomness.jpeg')
```

```python
import random
def polarization_string(dimension):
  l = [0,45,90,135]
  pol = random.choices(l,k=100)
  # You need to tell Python that if the bit in the string bit is 0,
  # then the key cannot be polarized in qubit 1, but can be in 0, + or -
  return pol
```

```python
random_pol_sequence = polarization_string(N)
fig=plt.figure(figsize=(20,3))
plt.plot(nbits,random_pol_sequence,marker='.',color='midnightblue',label='Alice polariz')
plt.title(' Random photon polarization')
plt.xlabel=('Number of bits')
plt.ylabel=('Value')
plt.legend(loc=4)
#plt.savefig('Random_photon_polarization.png')
```

```python
def random_polarization(key_bits,N):
  empty_list=[]
  lplus_minus=[45,135]
  lcomputational=[0,90]
  for i in range(N):
    if key_bits[i]==0:
      empty_list.append(random.choices(lcomputational,k=1))
    elif key_bits[i]==1:
      empty_list.append(random.choices(lplus_minus,k=1))
  connected_random_key = empty_list
  return connected_random_key

key_bits
```

```python
def Bob_guessing_key(N):
  l = [0,45,90,135]
  pol = random.choices(l,k=N)
  return pol 
Bob_guess = Bob_guessing_key(N)
Bob_gues = np.array(Bob_guess)
Bob_gues
```

```python
# Berlin calling
key_polarization = random_polarization(key_bits,N)
fig = plt.figure(figsize=(20,3))
plt.plot(nbits,key_polarization,marker='.',
         label='Randomness of Alice key polarization',color='midnightblue')
plt.plot(nbits,Bob_gues,marker='.',
         label='Randomness of Bob key polarization',color='orangered')
plt.xlabel=('Number of bits')
plt.ylabel=('Value')
plt.title(' Random photon polarization of Alice and Bob keys')
plt.legend(loc=4)
#plt.savefig('Comparison.jpeg')
```

```python
convergence_arr = np.zeros(100)
key_polarization = random_polarization(key_bits,N)
key_polarization=np.array(key_polarization)
for i in range(N):
  if key_polarization[i] == Bob_gues[i]:
    convergence_arr[i] = key_polarization[i]
  else:
    convergence_arr[i] = 0
fig = plt.figure(figsize=(20,3))
plt.bar(nbits,convergence_arr,color='blue')
unique, counts = np.unique(convergence_arr, return_counts=True)
dict(zip(unique, counts))
```
